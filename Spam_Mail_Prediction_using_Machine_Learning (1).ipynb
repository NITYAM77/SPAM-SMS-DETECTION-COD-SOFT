Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

The TfidfVectorizer is a feature extraction method provided by the sklearn.feature_extraction.text module in the scikit-learn library. It is used to convert a collection of raw text documents into a matrix of TF-IDF features.



Data Collection & Pre-Processing

```python
# loading the data from csv file to a pandas Dataframe
raw_mail_data = pd.read_csv("D:/IMP Notes and code/Machine Learning/Projects/17 Spam Mail Prediction/mail_data.csv")
```

```python
print(raw_mail_data)
```

Our dataset contain lot of null values so we first convert it into null string 

```python
# replace the null values with a null string
# This code is used for handling missing values in a pandas DataFrame (raw_mail_data). Specifically, it replaces any null (NaN) values in the
#  DataFrame with empty strings ('').



mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

Function: pd.notnull()<br>
This is a pandas function that returns a DataFrame of the same shape as the input but with boolean values.<br>
True indicates that the value is not null (i.e., it is a valid value).<br>
False indicates that the value is null (i.e., it is missing or NaN).

raw_mail_data.where(condition, other):<br>
It keeps the values in the DataFrame where the condition is True and replaces the values where the condition is False with the specified other value.

raw_mail_data.where((pd.notnull(raw_mail_data)), ''):<br>
This keeps all the non-null values in raw_mail_data as they are.<br>
It replaces all the null values in raw_mail_data with an empty string ('').

```python
# checking the number of rows and columns in the dataframe
mail_data.shape
```

Label Encoding

```python
# label spam mail as 0;  ham mail as 1;

mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1

# mail_data.loc[condition, 'Category'] selects the 'Category' column where the condition is True.
# mail_data.loc[mail_data['Category'] == 'spam', 'Category'] selects all rows in the 'Category' column where the 'Category' is 'spam'.
# = 0 assigns the value 0 to these selected rows.


# mail_data.loc[mail_data['Category'] == 'ham', 'Category'] selects all rows in the 'Category' column where the 'Category' is 'ham'.
# = 1 assigns the value 1 to these selected rows.

```

spam  -  0

ham  -  1

```python
# separating the data as texts and label

X = mail_data['Message']

Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

# min_df=1: This parameter means that terms which occur in at least one document will be included in the vocabulary. It helps to filter out rare words.
# stop_words='english': This parameter removes common English stop words (like 'and', 'the', etc.) which do not carry significant meaning.
# lowercase=True: This parameter converts all the characters in the text to lowercase, which ensures uniformity in the text data and helps to reduce the feature space.




X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# fit_transform: This method learns the vocabulary and inverse document frequency (IDF) from the training data and then transforms the training 
# data into a TF-IDF matrix.
# X_train: This is the input text data for training. It is typically a pandas Series or a list of strings.


# transform: This method transforms the test data into the same TF-IDF feature space learned from the training data. It uses the same vocabulary 
# and IDF values as the training data to ensure consistency.
# X_test: This is the input text data for testing. It is typically a pandas Series or a list of strings.


# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')

# astype('int'): This method converts the target variables (Y_train and Y_test) to integers. This is necessary because many machine learning 
# algorithms, including logistic regression, require the target labels to be numerical.
```

```python
print(X_train)
```

```python
print(X_train_features)
```

Training the Model

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with the training data
model.fit(X_train_features, Y_train)
```

Evaluating the trained model

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
# prediction on test data

prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

Building a Predictive System

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
